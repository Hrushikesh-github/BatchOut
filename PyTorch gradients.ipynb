Visualizing how .detach() method works. This is required to ensure the correct implementation of batchout

```python
import torch
from torchviz import make_dot
import numpy as np

x = torch.rand(3,2, requires_grad=True)
x = x ** 2
x = 5 * x
y = x.detach()

y = 5 * y
make_dot(y)
```

```python
make_dot(x) 
```

```python
print(x == y)
x.data = y # Try x = y, we will get wrong/other result
print(x == y)

print(x.requires_grad)

make_dot(x)

```
